```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import plot, savefig

labels = ['Female', 'Male']
sizes = [28, 72]
explode = (0.1, 0)
colors = ['#f28a90', '#71afe2']
fig, ax = plt.subplots()
ax.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%.0f%%', shadow = True, startangle = 90)
ax.axis('equal')
plt.title('Gender Gap in STEM in 2023', y = -0.1)

savefig('viz1.png', transparent = True, dpi = 300)
```

```python
import numpy as np

labels = ['Female', 'Male']
sizes = [60, 46]
percent = ['60%', '46%']
colors = ['#f28a90', '#71afe2']

fig = plt.figure(figsize = (7, 5))
fig, ax = plt.subplots()
ax.set_ylim(ymin = 0, ymax = 100)
bp1 = plt.bar(labels, sizes, color = colors, width = 0.4)
plt.bar_label(bp1, labels = percent, label_type = 'edge', padding = 3)
plt.xlabel('Gender Breakdown')
plt.ylabel('% of People in STEM')
plt.title('People in STEM inspired by a Role Model')

savefig('viz2.png', dpi = 300)
```
